# Logistic Regression 
used for classification and predictive analytics. Logistic regression estimates the probability of an event occurring,such as voted or didn't vote, based on a given dataset of independent variables

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_predict=lr.predict(x_test)
result=classification_report(y_test,lr_predict)
result1=confusion_matrix(y_test,lr_predict)
print(result)
```

```python
print(result1)
```

```python
sns.heatmap(confusion_matrix(y_test,lr_predict), annot = True, fmt = "d")
```

```python
def performance_metrices(y_testing,pred):
    confusion_mat = confusion_matrix(y_testing, pred)
    TP = confusion_mat[0,0:1]
    FP = confusion_mat[0,1:2]
    FN = confusion_mat[1,0:1]
    TN = confusion_mat[1,1:2]
    
    accuracy = (TP+TN)/((FN+FP)+(TP+TN))
    sensitivity = TP/(TP+FN)
    specificity = TN/(TN+FP)
    precision = TP/(TP+FP)
    recall =  TP / (TP + FN)
    fScore = (2 * recall * precision) / (recall + precision)
    
    print("Confusion Matrix:\n",confusion_mat)
    print("Accuracy :",accuracy)
    print("Sensitivity :",sensitivity)
    print("Specificity :",specificity)
    print("Precision:",precision)
    print("Recall:",recall)
    print("F-score:",fScore)
```

```python
performance_metrices(y_test,lr_predict)
```

# AUC(Area under ROC curve)
measures the entire two-dimensional area underneath the entire ROC curve (think integral calculus) from (0,0) to (1,1)

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(y_test,lr_predict)

auc_score = roc_auc_score(y_test,lr_predict)
fpr
tpr

plt.plot(fpr, tpr, label=f"AUC = {auc_score:.2f}")
plt.plot([0, 1], [0, 1], 'r--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend()
plt.show()
#ROC Curve for the predictions made by a logistic regression model (lr_predict).
#The roc_curve and roc_auc_score functions from scikit-learn are used to compute the necessary metrics.

#The plt.plot lines visualize the ROC Curve, and the plt.title, plt.xlabel, and plt.ylabel functions add labels to the plot.
#The legend displays the AUC score.
```
